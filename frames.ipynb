```python
# Read the video from specified path 
def create_frames(file):
    print("own_dataset/d6/"+file)
    cam = cv2.VideoCapture("own_dataset/d6/"+file) 
    # 5 for the train and 1 for the test
    tt = 5*["train"]+["test"]
    random.shuffle(tt)
    
    # Define the shape to save the different frames

    

    # frame 
    currentframe = 0

    files=[]

    while(True): 

        # reading from frame 
        ret,img = cam.read() 
        
    

        if ret: 
            M = img.shape[0]//3  # cols
            N = img.shape[1]//2  # rows
            # if video is still left continue creating images 
            tiles = [img[c:c+M,r:r+N] for c in range(0,img.shape[0],M) for r in range(0,img.shape[1],N)]

            for i in range(len(tiles)):
                name = "own_dataset/d6/"+tt[i]+"/"+ file[1:-4]+"_"+str(currentframe)+"_"+str(i+1)+ '.jpg'
            
                print ('Creating...' + name) 
                files.append(name)
                # writing the extracted images 
                cv2.imwrite(name, tiles[i]) 

            # increasing counter so that it will 
            # show how many frames are created 
            currentframe += 1
        else: 
            break

    # Release all space and windows once done 
    cam.release() 
    cv2.destroyAllWindows() 
    return files
```

```python
frames=[]
for v in videos:
    temp=create_frames(v)
    for f in temp:
        frames.append(f)
```
